## Analisis de que predijo Keras y que XGBoost

```python
pred_keras = pd.read_csv("../data/submit_keras_codificado.csv")
```

```python
pred_xgb = pd.read_csv("../data/submit_xgb_codificado_reducido.csv")
```

```python
comparacion = pred_keras.copy()
```

```python
comparacion["xgb"] = pred_xgb["SCORE"]
comparacion.rename(columns = {"SCORE":"Keras"},inplace = True)
```

```python
comparacion["Ensamble_50_50"] = comparacion["Keras"] * 0.5 + comparacion["xgb"] * 0.5
```

```python
comparacion.head()
```

```python
submit = comparacion[["USER_ID","Ensamble_50_50"]].rename(columns = {"Ensamble_50_50": "SCORE"})
```

```python
submit.set_index("USER_ID",inplace = True)
```

```python
submit.to_csv("../data/ensamble_submission.csv")
```
